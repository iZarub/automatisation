```python
import pyvisa as visa
import matplotlib.pyplot as plt
import csv
import time
import numpy as np
import datetime
import tqdm


rm = visa.ResourceManager()
osc = rm.open_resource('ASRL3::INSTR')
osc.read_termination = '\n'
osc.write_termination = '\n'

#osc.clear()

#используем всю память на точки
osc.write('ACQ:WRAT MSAM')
time.sleep(1)

#фказываем количество точек: DEF - на экране, MAX - миллион
osc.write('POD:DATA:POIN MAX')
time.sleep(1)
osc.write('CHAN2:DATA:POIN MAX')
time.sleep(1)
osc.write('CHAN1:DATA:POIN MAX')
time.sleep(1)

#включаем каналы
osc.write('CHAN2:STAT ON') #on and off chanels
osc.write('CHAN1:STAT ON')

for i in range(60):
    #очистка
    #osc.clear()
    name = "7_7_22/FILE" + str(i)
    f = open(name, 'w')
    f.write(str(datetime.datetime.now()) + '\n')
    
    ts = time.time()
    
    #фиксируем
    osc.write('SING')
    time.sleep(1)


    #фересчет времени
    timeinfo_str = osc.query('CHAN2:DATA:HEAD?').split(',')
    print(timeinfo_str)
    timeinfo_fl = []
    for elem in timeinfo_str:
        timeinfo_fl.append(float(elem))
    dt = (timeinfo_fl[1] - timeinfo_fl[0])/timeinfo_fl[2]
    timeaxis = []
    for i in range(int(timeinfo_fl[2])):
        timeaxis.append(timeinfo_fl[0] + i*dt)
        
    time.sleep(1)

    #снимаем
    query2 = osc.query('CHAN2:DATA?').split(',')
    trace2 = []
    for elem in query2:
        try:
            trace2.append(float(elem))
        except:
            print(elem)
            
            

    query1 = osc.query('CHAN1:DATA?').split(',')
    trace1 = []
    for elem in query1:
        try:
            trace1.append(float(elem))
        except:
            print(elem)

    te = time.time()
    print('Aqusition took %.1f seconds' % float(te-ts))
    
    for i in range(len(trace1)):
        f.write(str(timeaxis[i])+','+str(trace1[i])+','+str(trace2[i]) + '\n')
    
    f.close()


osc.close()
rm.close()
```
