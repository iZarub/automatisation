```python
import pyvisa as visa
import matplotlib.pyplot as plt
import csv
import time
import numpy as np

rm = visa.ResourceManager()
pm = rm.open_resource('USB0::0x1313::0x8078::P0034218::INSTR')
pm.read_termination = '\n'
pm.write_termination = '\n'
pm.timeout = 1000
time.sleep(1)

pm.clear()


pm.write('CONF:POW')
print(pm.query('CONF?'))
pm.write('MEAS:POW')
print(pm.query('READ?'))
#print(pm.query('FETC?'))

pm.close()
rm.close()
```
